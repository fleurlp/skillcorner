```python
# new data set without the Middlesbrough players
data1 = data[data['Team'] != 'Middlesbrough']
data1 = data1.rename(columns={'Count Performances (PQC passed)': 'Count Performances'})
```

```python
variables = ['Minutes Played',
             'Count Performances',
             'Distance',
             'M/min',
             'Sprinting Distance',
             'HI Distance',
             'Count Sprint',
             'Count HI',
             'PSV-99',
             'Minutes Played TIP', 
             'Distance TIP', 
             'M/min TIP', 
             'Sprinting Distance TIP',
             'HI Distance TIP',
             'Count Sprint TIP',
             'Count HI TIP',
             'Minutes Played OTIP',
             'Distance OTIP', 
             'M/min OTIP', 
             'Sprinting Distance OTIP',
             'HI Distance OTIP',
             'Count Sprint OTIP',
             'Count HI OTIP',
          ]
```

```python
# calculating the mean of each variable for all the players in each league and then ranking the leagues 
sorted_data = {}

for var in variables:
    grouped_data = data1.groupby('Competition')[var].mean()
    sorted_data[var] = grouped_data.sort_values(ascending=False)

for var, sdata in sorted_data.items():
    print(f"Rankings for {var}:")
    print(sdata)
    print("\n" + "="*50 + "\n")  # for better readability
```

```python
# retrieving top 5 players for each variable in top league (ENG)
ENGdata = data1[data1['Competition'] == 'ENG - Championship']

for metric in variables:
    top_players_ENG = ENGdata.nlargest(5, metric)[['Player', 'Competition']]
    print(f"Top 5 ENG Players for {metric}:")
    print(top_players_ENG)
    print("\n" + "="*50 + "\n")  # for better readability
```

```python
# calculating how many times each player makes the top 5 across all variables (ENG)
player_counts_ENG = {}

for metric in variables:
    top_players_ENG = ENGdata.nlargest(5, metric)['Player'].tolist()
    
    for player in top_players_ENG:
        if player in player_counts_ENG:
            player_counts_ENG[player] += 1
        else:
            player_counts_ENG[player] = 1

sorted_players_ENG = sorted(player_counts_ENG.items(), key=lambda x: x[1], reverse=True)

for player, count in sorted_players_ENG:
    print(f"{player}: {count}")
```

```python
# retrieving top 5 players for each variable in all leagues 
for metric in variables:
    top_players_all5 = data1.nlargest(5, metric)[['Player', 'Competition']]
    print(f"Top 5 All Leagues Players for {metric}:")
    print(top_players_all5)
    print("\n" + "="*50 + "\n")  # for better readability
```

```python
# calculating how many times each player makes the top 5 across all variables (ALL LEAGUES)
player_counts_all5 = {}
player_competition_all5 = {}

for metric in variables:
    top_players_all5 = data1.nlargest(5, metric)[['Player', 'Competition']].values
    
    for player, competition in top_players_all5:
        if player in player_counts_all5:
            player_counts_all5[player] += 1
        else:
            player_counts_all5[player] = 1
        
        player_competition_all5[player] = competition

sorted_players_all5 = sorted(player_counts_all5.items(), key=lambda x: x[1], reverse=True)

print("Players making top 5 across all variables:\n")
for player, count in sorted_players_all5:
    print(f"{player}, {player_competition_all5[player]}: {count}")
    
    

# calculating how many times each league makes the top 5 across all variables
competition_counts_all5 = {}

for metric in variables:
    top_competitions_all5 = data1.nlargest(5, metric)['Competition'].tolist()
    
    for competition in top_competitions_all5:
        if competition in competition_counts_all5:
            competition_counts_all5[competition] += 1
        else:
            competition_counts_all5[competition] = 1

sorted_competitions_all5 = sorted(competition_counts_all5.items(), key=lambda x: x[1], reverse=True)

print("\n\nLeagues associated with each top 5 players across all vairables:\n")
for competition, count in sorted_competitions_all5:
    print(f"{competition}: {count}")
```

```python
# calculating how many times each player makes the top player across all variables (ALL LEAGUES)
player_counts_all1 = {}
player_competition_all1 = {}

for metric in variables:
    top_players_all1 = data1.nlargest(1, metric)[['Player', 'Competition']].values
    
    for player, competition in top_players_all1:
        if player in player_counts_all1:
            player_counts_all1[player] += 1
        else:
            player_counts_all1[player] = 1
        
        player_competition_all1[player] = competition

sorted_players_all1 = sorted(player_counts_all1.items(), key=lambda x: x[1], reverse=True)

print("Players making top player across all variables:\n")
for player, count in sorted_players_all1:
    print(f"{player}, {player_competition_all1[player]}: {count}")
    
    

# calculating how many times each league makes the top 1 across all variables
competition_counts_all1 = {}

for metric in variables:
    top_competitions_all1 = data1.nlargest(1, metric)['Competition'].tolist()
    
    for competition in top_competitions_all1:
        if competition in competition_counts_all1:
            competition_counts_all1[competition] += 1
        else:
            competition_counts_all1[competition] = 1

sorted_competitions_all1 = sorted(competition_counts_all1.items(), key=lambda x: x[1], reverse=True)

print("\n\nLeagues associated with each top player across all vairables:\n")
for competition, count in sorted_competitions_all1:
    print(f"{competition}: {count}")
```

```python
# isolating the Middlesbrough players
MIDdata = data[data['Team'] == 'Middlesbrough']
```

```python
# isolating the top 5 players from top league (ENG)
top5_players_ENG = sorted_players_ENG[:5]
```

```python
# comparison between each top 5 ENG player and each middlesbrough player (50%)
top5_ENG_names = [player[0] for player in top5_players_ENG]

# creating a dictionary to store comparisons
comparisons_ENG = {}
replacement_candidates_ENG = {}

for MID_player in MIDdata['Player'].unique():
    comparisons_ENG[MID_player] = {}
    for ENG_player in top5_ENG_names:
        comparisons_ENG[MID_player][ENG_player] = {}
        
        ENG_higher_count = 0

        for metric in variables:
            MID_value = MIDdata[MIDdata['Player'] == MID_player][metric].values[0]
            ENG_value = data[data['Player'] == ENG_player][metric].values[0]
            
            if ENG_value > MID_value:
                ENG_higher_count += 1

            comparisons_ENG[MID_player][ENG_player][metric] = (MID_value, ENG_value)

        # if more than half of the ENG player's variables are higher, replace
        if ENG_higher_count > (1/2) * len(variables):
            if MID_player in replacement_candidates_ENG:
                replacement_candidates_ENG[MID_player].append(ENG_player)
            else:
                replacement_candidates_ENG[MID_player] = [ENG_player]

for MID_player, ENG_players in replacement_candidates_ENG.items():
    for ENG_player in ENG_players:
        print(f"{MID_player} should be replaced by {ENG_player}")
```

```python
# comparison between each top 5 ENG player and each middlesbrough player (75%)

# creating a dictionary to store comparisons
comparisons_ENG1 = {}
replacement_candidates_ENG1 = {}

for MID_player in MIDdata['Player'].unique():
    comparisons_ENG1[MID_player] = {}
    for ENG1_player in top5_ENG_names:
        comparisons_ENG1[MID_player][ENG1_player] = {}
        
        ENG1_higher_count = 0

        for metric in variables:
            MID_value = MIDdata[MIDdata['Player'] == MID_player][metric].values[0]
            ENG1_value = data[data['Player'] == ENG1_player][metric].values[0]
            
            if ENG1_value > MID_value:
                ENG1_higher_count += 1

            comparisons_ENG1[MID_player][ENG1_player][metric] = (MID_value, ENG1_value)

        # if more than three quarters of the ENG player's variables are higher, replace
        if ENG1_higher_count > (3/4) * len(variables):
            if MID_player in replacement_candidates_ENG1:
                replacement_candidates_ENG1[MID_player].append(ENG1_player)
            else:
                replacement_candidates_ENG1[MID_player] = [ENG1_player]

for MID_player, ENG1_players in replacement_candidates_ENG1.items():
    for ENG1_player in ENG1_players:
        print(f"{MID_player} should be replaced by {ENG1_player}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# excluding distance variables to make plots more digestible  
exclude_vars = ['Distance', 'Sprinting Distance', 'HI Distance', 'Distance TIP', 'Sprinting Distance TIP', 'HI Distance TIP', 'Distance OTIP', 'Sprinting Distance OTIP', 'HI Distance OTIP']

# remaining variables
remaining_vars = [var for var in variables if var not in exclude_vars]

# for each Middlesbrough player to be replaced
for MID_player, ENG1_players in replacement_candidates_ENG1.items():
    for ENG1_player in ENG1_players:

        # values for the main plot
        MID_values_main = [MIDdata[MIDdata['Player'] == MID_player][metric].values[0] for metric in remaining_vars]
        ENG1_values_main = [data[data['Player'] == ENG1_player][metric].values[0] for metric in remaining_vars]

        # values for the distance plot
        MID_values_dist = [MIDdata[MIDdata['Player'] == MID_player][metric].values[0] for metric in exclude_vars]
        ENG1_values_dist = [data[data['Player'] == ENG1_player][metric].values[0] for metric in exclude_vars]

        ENG1_team = data[data['Player'] == ENG1_player]['Team'].values[0]

        bar_width = 0.35
        index_main = np.arange(len(remaining_vars))
        index_dist = np.arange(len(exclude_vars))

        # create a figure with two subplots side by side
        fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(20, 8))

        # plot for the main variables on the first subplot
        ax1.bar(index_main, MID_values_main, bar_width, label=f"{MID_player} (Middlesbrough)", color='b', alpha=0.7)
        ax1.bar(index_main + bar_width, ENG1_values_main, bar_width, label=f"{ENG1_player} ({ENG1_team})", color='r', alpha=0.7)
        ax1.set_xlabel('Variables')
        ax1.set_ylabel('Values')
        ax1.set_title(f'Comparison between {MID_player} and {ENG1_player}')
        ax1.set_xticks(index_main + bar_width / 2)
        ax1.set_xticklabels(remaining_vars, rotation=90)
        ax1.legend()

        # plot for the distance variables on the second subplot
        ax2.bar(index_dist, MID_values_dist, bar_width, label=f"{MID_player} (Middlesbrough)", color='b', alpha=0.7)
        ax2.bar(index_dist + bar_width, ENG1_values_dist, bar_width, label=f"{ENG1_player} ({ENG1_team})", color='r', alpha=0.7)
        ax2.set_xlabel('Variables')
        ax2.set_ylabel('Values')
        ax2.set_title(f'Comparison (Distance Variables) between {MID_player} and {ENG1_player}')
        ax2.set_xticks(index_dist + bar_width / 2)
        ax2.set_xticklabels(exclude_vars, rotation=90)
        ax2.legend()

        plt.tight_layout()
        plt.show()
```
